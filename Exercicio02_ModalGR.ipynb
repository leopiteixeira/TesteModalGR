```python
import pandas as pd
from datetime import date
```

```python
dados = pd.read_csv('teste.txt', sep="|")
dados
```

```python
dadosAniversariantes = dados.copy()
```

```python
for id, data in enumerate(dados["DtNascimento"]):
  if int(data.split("-")[1]) != date.today().month:
    dadosAniversariantes = dadosAniversariantes.drop(id)
dadosAniversariantes = dadosAniversariantes.reset_index().drop("index", axis=1)
```

```python
dadosAniversariantes
```

```python
arq = open("aniversariantes.txt", 'w')
arq.write("Nome|Email|DtNascimento\n")
for id in range(len(dadosAniversariantes)):
  pessoa = dadosAniversariantes.iloc[id]
  arq.write(f"{pessoa.Nome}|{pessoa.Email}|{pessoa.DtNascimento}\n")
arq.close()
```
